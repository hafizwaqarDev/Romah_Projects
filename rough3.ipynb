```python
header ={
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36'
}
data_list = []
for link in allAgentlink:
    time.sleep(1)
    pageResponse = requests.get(link,headers=header)
    # if pageResponse.status_code == 200:
    pageSoupTree = fromstring(pageResponse.text)
    data = pageSoupTree.find('.//script[@id="__NEXT_DATA__"]').text_content()
    try:
        parsed_data = json.loads(data)    
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")
    allData = parsed_data['props']['pageProps']['page']['propertiesByFilters']['properties']
    try: totalPropties = parsed_data['props']['pageProps']['page']['propertiesByFilters']['paginationResponse']['totalCount']
    except: totalPropties = ''
    for data in allData:
        try: title = data['title']
        except: title = ''
        try: propertyType = data['priceType']['formattedValue']
        except: propertyType = ''
        try: propertyCataGeory = data['propertyType']['formattedValue']
        except: propertyCataGeory = ''
        try: price = data['price']['display']
        except: price = ''
        try: propertyUrl = 'https://www.rumah123.com' + data['url']
        except: propertyUrl = ''
        try: fullAddress = data['location']
        except: fullAddress = ''
        try: city = data['location']['city']['name']
        except: city = ''
        try: district = data['location']['district']['name']
        except: district = ''
        try: province = data['location']['province']['name']
        except: province = ''
        try:
            Address = city+district+province
        except: Address = ''
        agentType = ''
        agencyname = ''
        agentName = ''
        ownerName = ''
        try: agencyName = data['agent']['organization']['name']
        except: agentName = ''
        if agencyName:
            agentName = data['agent']['name'].strip()
            AgentType = 'Agen Korporat'
        else:
            AgentType = 'Agen Independen'
            ownerName = data['agent']['name'].strip()
        # try: agentLink = data['agent']['contacts'][0]['value']
        # except: agentLink = ''
        try: email = data['agent']['contacts'][1]['value']
        except: email = ''
        try: phoneNumber = data['agent']['contacts'][2]['value']
        except: phoneNumber = ''
        try: WhatsAppNum = data['agent']['contacts'][3]['value']
        except: WhatsAppNum = ''
        row = [title,propertyUrl,price,propertyType,propertyCataGeory,city,district,province,Address,link,agencyName,AgentType,agentName,ownerName,phoneNumber,WhatsAppNum,email,totalPropties]
        print(row)
        with open(file='rumah1233.csv',mode='a',newline='',encoding='UTF-8') as file:
            csv.writer(file).writerow(row)
```

```python
import pandas as pd
header ={
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36'
}
data_list = []

for page in range(1, 6):
    webURl = f'https://www.rumah123.com/perumahan-baru/jawa-timur/residensial/?page={page}'
    pageresponse1 = requests.get(webURl, headers=header)
    time.sleep(6)
    htmlContent = pageresponse1.text
    souptree = fromstring(htmlContent)
    dataa = souptree.find('.//script[@type="text/babel"]').text_content()
    try:
        parsed_data = json.loads(dataa)
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")
    alldata = parsed_data['search']['ui']['cardList']
    for data in alldata:
        try:
            title = data['card']['title'].strip()
        except:
            title = ''
        try: link = 'https://www.rumah123.com' + data['card']['link']
        except: link = ''
        try:  address = data['card']['address']
        except: address = ''
        splittedAddress = address.split(',')
        city,province = splittedAddress[0].strip(),splittedAddress[1].strip()
        try:
            shortDescription = data['card']['shortDescription'].strip()
        except:
            shortDescription = ''
        try: propertyCatageory = data['card']['propertyType']
        except: propertyCatageory = ''
        try: propertyType = data['card']['priceType']
        except: propertyType = ''
        try: price = data['card']['price']['display']
        except: price = ''
        agencyName = ''
        agentName = ''
        AgentType = ''
        ownerName = ''
        emailTag = ''
        email = ''
        try: agencyName = data['card']['agentOrganizationName']
        except: agencyName = '' 
        if agencyName:
            agentName = data['card']['agencyName']
            AgentType = 'Agen Korporat'
        else:
            AgentType = 'Agen Independen'
            ownerName = data['card']['agencyName']
        try: phoneNumber = data['card']['phoneNumber']
        except: phoneNumber = ''
        try: whatsappNumbers = data['card']['whatsappNumbers'][0]
        except: whatsappNumbers = ''
        try: emailTag = data['card']['agent']['contacts']['1']['value'][0]
        except: emailTag = ''
        if '@' in emailTag:
            email = emailTag
        try: agentLink = data['card']['agent']['contacts']['5']['value'][0]
        except: agentLink = ''
        row = [title, link, price, propertyType, propertyCatageory, city, province, address, shortDescription, agentLink, AgentType, agencyName, agentName, ownerName, phoneNumber, whatsappNumbers, email]
        print(row)
        data_list.append(row)

df = pd.DataFrame(data_list, columns=['Property Title','Property Url','Property Price','Property Type','Property Catageory','City','Province','Property Complete Address','Property Description','Agent Link','Agent Type','Agency Name','Agent Name','Owner Name','Phone Number','Whatsapp Number','Email'])

try:
    existing_df = pd.read_excel('rumah123_data.xlsx')
except FileNotFoundError:
    existing_df = pd.DataFrame(columns=['Property Title','Property Url','Property Price','Property Type','Property Catageory','City','Province','Property Complete Address','Property Description','Agent Link','Agent Type','Agency Name','Agent Name','Owner Name','Phone Number','Whatsapp Number','Email'])
updated_df = pd.concat([existing_df, df], ignore_index=True)
updated_df.to_excel('rumah123_data.xlsx', index=False)
```
